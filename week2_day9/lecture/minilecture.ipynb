## Minilecture on Recursion

Taught by AddisCoder TAs in Amharmic in Week 2, Day 3 Morning lab.

Talking points:
 - "Recursion" is a word that means a function calling itself.
 - Two important parts of a recursive solution:
   - A "Base Case", which is the simplist version of the problem to which we have a solution.
   - A "Recursive Case", which is the general form of the problem, in which you make some progress towards the base case and call the same function with a simpler input.
- The "trick" in writing recursive functions is to recognize how to state your problem in a way that it refers to itself.
- This is hard at first, but gets much easier with practice!

Let's go through a practice problem together:
 - pick 1 simple recursive problem statement (counting up from 1 to n, for example) and show how to recognize the base case and the recursive case.
